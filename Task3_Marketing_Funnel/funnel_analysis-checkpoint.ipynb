```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Channel": ["Social Media", "Google Ads", "Email"],
    "Visitors": [1000, 1200, 800],
    "Leads": [300, 400, 200],
    "Conversions": [100, 150, 80]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Channel", y="Conversions", data=df)
plt.title("Conversions by Channel")
plt.show()
```

```python
df["Conversion Rate (%)"] = (df["Conversions"] / df["Visitors"]) * 100
df
```

```python
df["Drop Visitors→Leads"] = df["Visitors"] - df["Leads"]
df["Drop Leads→Conversions"] = df["Leads"] - df["Conversions"]

df
```
